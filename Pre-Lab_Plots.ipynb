## Partial Decay Width Calculation (Q 1)

```python
G_F = 1.1663787E-5 #Fermi's constant
M_Z = 91.182 #Mass of the Z Boson
angle_wein = 0.2312 #Sine squared of the Weinberg angle
```

```python
#Variables
N = 3 #Color factor
Q_f = 2/3 #Charge of the Fermion in unis of the electron's charge
I_v = -0.5 #Axial vector coupling
G_f_v = I_v -2*Q_f*angle_wein
```

```python
Partial_width = (mth.sqrt(2)/(12 * mth.pi))*N*G_F*(M_Z**3)*((G_f_v**2)+(I_v**2))
display(Partial_width*1000)
```

### Angular distribution plot for the process $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$

```python
#The x-axis values
x = np.linspace(-1.0, 1.0, 10000)

#Corresponding y-axis values
y = (x**2) + 1 

#Plotting the points 
plt.plot(x, y, label=r'$s$-channel', color='#0a55a1')
  
#Labeling the x axis
plt.xlabel(r'$\cos{(\theta)}$')

#Labeling the y axis
plt.ylabel(r'$\frac{d \sigma}{d \Omega}$')

#Titling the plot
plt.title(r'Angular distribution for the process: $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$', size = 13)

plt.xticks(np.arange(-1.0, 1.5, 0.5))
plt.legend(loc='upper right')
plt.show()
```

## Calculating $A_{FB}$ for the $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$ process (Q 5.5)

```python
#Setting constants

T_Z = 2.4862 #
Q = -1 #Charge of the Fermion in unis of the electron's charge
I = -0.5 #Third component of weak isospin
s = 91.225**2 #Center of mass Energy squared
```

```python

sin_theta=mth.sqrt(0.23)
cos_theta=mth.sqrt(1-(sin_theta**2))

g_v=I-2*(Q*(sin_theta**2)) 

v_f=g_v/(2*sin_theta*cos_theta)

a_f=I/(2*sin_theta*cos_theta)

prop_denom = ((s-(M_Z**2))**2)+((s*T_Z/M_Z)**2)

Re_chi=(s*(s-(M_Z**2)))/prop_denom #Real part of the propagator

mod_chi=(s**2)/prop_denom #Modulus of the propagator

F_1=(Q**2)-(2*(v_f**2)*Q*Re_chi)+((((v_f**2)+(a_f**2))**2)*mod_chi)
F_2=(-2*(a_f**2)*Q*Re_chi)+(4*(v_f**2)*(a_f**2)*mod_chi)

N_c=1.0
```

```python
#Calculating the forward backward assymetry
A_FB=(3/4)*(F_2/F_1)
display(A_FB)
```
