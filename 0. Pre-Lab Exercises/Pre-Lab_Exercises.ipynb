## 5.1 Partial Decay Width Calculation 

```python
G_F = 1.1663787E-5 #Fermi's constant
M_Z = 91.182 #Mass of the Z Boson
angle_wein = 0.2312 #Sine squared of the Weinberg angle
```

```python
#Variables
N = 1 #Color factor
Q_f = 0 #Charge of the Fermion in unis of the electron's charge
I_v = 0.5 #Axial vector coupling
G_f_v = I_v -2*Q_f*angle_wein
```

```python
Partial_width = (mth.sqrt(2)/(12 * mth.pi))*N*G_F*(M_Z**3)*((G_f_v**2)+(I_v**2))*1000
display(Partial_width)
```

## 5.2 Decay width and cross section calculations

```python
electron_width = 83.39496752776024
neutrino_width = 165.8520351625946
neutral_leptonic_width = 3 *neutrino_width
charged_leptonic_width = 3 * electron_width
hadronic_width = (2 * 285.3600405624264) + (3 * 367.8171081972608)
total_width = hadronic_width + charged_leptonic_width + neutral_leptonic_width
```

```python
Partial_width = total_width
display(Partial_width)
```

```python
partial_cross_section = (12 * mth.pi * electron_width * Partial_width)/((M_Z**2)*(total_width**2))
display(partial_cross_section*100000)
```

## 5.3 Percentage increase in the width of the $Z^0$ resonance curve

```python
new_total_width = total_width + electron_width + neutrino_width + 285.3600405624264 + 367.8171081972608
display(new_total_width)
```

```python
percentage_inc = ((new_total_width-total_width)/total_width) * 100
display(percentage_inc)
```

### 5.4.1 Angular distribution plot for the process $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$

```python
#The x-axis values
x = np.linspace(-1.0, 1.0, 10000)

#Corresponding y-axis values
y = (x**2) + 1 

#Plotting the points 
plt.plot(x, y, label=r'$s$-channel', color='#0a55a1')
  
#Labeling the x axis
plt.xlabel(r'$\cos{(\theta)}$')

#Labeling the y axis
plt.ylabel(r'$\frac{d \sigma}{d \Omega}$')

#Titling the plot
plt.title(r'Angular distribution for the process: $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$', size = 13)

plt.xticks(np.arange(-1.0, 1.5, 0.5))
plt.legend(loc='upper right')
plt.savefig('ee_mm.png')
plt.show()
```

## 5.5 Calculating $A_{FB}$ for the $e^{+}e^{-} \rightarrow \mu^{+}\mu^{-}$ process

```python
#Setting constants

T_Z = 2.4862 #
Q = -1 #Charge of the Fermion in unis of the electron's charge
I = -0.5 #Third component of weak isospin
s = 91.225**2 #Center of mass Energy squared
```

```python

sin_theta=mth.sqrt(0.23)
cos_theta=mth.sqrt(1-(sin_theta**2))

g_v=I-2*(Q*(sin_theta**2)) 

v_f=g_v/(2*sin_theta*cos_theta)

a_f=I/(2*sin_theta*cos_theta)

prop_denom = ((s-(M_Z**2))**2)+((s*T_Z/M_Z)**2)

Re_chi=(s*(s-(M_Z**2)))/prop_denom #Real part of the propagator

mod_chi=(s**2)/prop_denom #Modulus of the propagator

F_1=(Q**2)-(2*(v_f**2)*Q*Re_chi)+((((v_f**2)+(a_f**2))**2)*mod_chi)
F_2=(-2*(a_f**2)*Q*Re_chi)+(4*(v_f**2)*(a_f**2)*mod_chi)

N_c=1.0
```

```python
#Calculating the forward backward assymetry
A_FB=(3/4)*(F_2/F_1)
display(A_FB)
```
